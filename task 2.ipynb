```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import seaborn as sns
```

```python
# Load data from CSV file using pandas module
def load_data(file_path):
    return pd.read_csv(file_path)
```

```python
data = load_data('Airbnb_UK_2022.csv')
```

```python
# Display the proportion of number of bedrooms of Airbnb listing using pie chart
bedroom_count = data['bedrooms'].value_counts()
plt.pie(bedroom_count, labels=bedroom_count.index, autopct='%1.1f%%')
plt.title('Proportion of Number of Bedrooms')
plt.show()
```

```python
# Display the number of listings for each room type using bar chart
room_type_count = data['room_type'].value_counts()
plt.bar(room_type_count.index, room_type_count)
plt.title('Number of Listings for Each Room Type')
plt.xlabel('Room Type')
plt.ylabel('Number of Listings')
plt.show()
```

```python
# Display the relationship between accommodates and price using scatter plot
plt.scatter(data['accommodates'], data['price'])
plt.title('Relationship between Accommodates and Price')
plt.xlabel('Accommodates')
plt.ylabel('Price')
plt.show()
```

```python
# Convert date column to proper format
data['host_since'] = pd.to_datetime(data['host_since'])
```

```python
# Display Airbnb prices from 2019 - 2022 with line chart using subplots (one year per plot)
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
years = [2019, 2020, 2021, 2022]
for i, year in enumerate(years):
    year_data = data[data['host_since'].dt.year == year]
    row = i // 2
    col = i % 2
    axs[row, col].plot(year_data['host_since'], year_data['price'])
    axs[row, col].set_title(f'Airbnb Prices in {year}')
    axs[row, col].set_xlabel('Date')
    axs[row, col].set_ylabel('Price')

plt.show()
```

```python
# Display a visualisation of your choice to present the information from the Airbnb services or properties that can indicate customer behaviour or pattern
# Example: Distribution of review scores by room type
plt.figure(figsize=(12,6))
sns.boxplot(x='room_type', y='review_scores_rating', data=data)
plt.title('Distribution of Review Scores by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Review Scores')
plt.show()
```
